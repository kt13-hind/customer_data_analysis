```python
import geopandas as gpd

# 1. Read the Shapefile
map_df = gpd.read_file('Admin2.shp')
```

```python
map_df.head()
```

```python
# the columns ST_NM and ship-state have different strings at some states
# then there is also the matter of some missing territory
# nope there is zero indexing in the columns so its clear that there are 36 total
map_df['ST_NM'] = map_df['ST_NM'].str.lower()
cancelled_orders_statewise['ship-state'] = cancelled_orders_statewise['ship-state'].str.lower()
def difference_between_states():
    map_df_state = set(map_df['ST_NM'].unique())
    cancelled_orders_statewise_state = set(cancelled_orders_statewise['ship-state'].unique())
    return map_df_state ^ cancelled_orders_statewise_state
```

```python
unique_to_both = difference_between_states()
print(unique_to_both)
map_df.loc[map_df["ST_NM"] == "dadra and nagar haveli and daman and diu", "ST_NM"] = 'dadra and nagar'
map_df.loc[map_df["ST_NM"] == "delhi", "ST_NM"] = 'new delhi'
map_df.loc[map_df["ST_NM"] == "andaman & nicobar", "ST_NM"] = 'andaman & nicobar '
```

```python
unique_to_both = difference_between_states()
print(unique_to_both)
```

```python

# 3. Merge dataframes (map geometry + analysis data)
merged = map_df.set_index('ST_NM').join(cancelled_orders_statewise.set_index('ship-state'))

# 4. Create the map
fig, ax = plt.subplots(1, figsize=(10, 12))
merged.plot(column='count', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Analysis of Cancelled orders in Indian States', fontsize=15)
plt.show()
```
